#### Thay đổi kích thước ảnh

```python
def rescaleImg(image, scale = 0.5):
  width = int(image.shape[1] * scale)
  height = int(image.shape[0] * scale)
  dims = (width, height)
  if (scale <= 0):
    print("Invalid inputting")
  elif (scale <= 1):
    return cv.resize(image, dims, interpolation = cv.INTER_AREA)
  else:
    return cv.resize(image, dims, interpolation = cv.INTER_LINEAR)
```

Thu nhỏ ảnh:

```python
small_gray = rescaleImg(gray)
```

```python
cv2_imshow(small_gray)
```

Phóng to ảnh:

```python
massive_gray = rescaleImg(gray, scale = 1.5)
```

```python
cv2_imshow(massive_gray)
```
